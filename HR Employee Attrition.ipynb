```python
mean_age = df['ï»¿Age'].mean()
print(f"Mean Age: {mean_age:.2f}")
```

```python
job_satisfaction_distribution = df['JobSatisfaction'].value_counts()
print("Job Satisfaction Distribution:")
print(job_satisfaction_distribution)
```

```python
job_level_monthly_income = df.groupby('JobLevel')['MonthlyIncome'].mean()
print("Job Level vs. Monthly Income:")
print(job_level_monthly_income)
```

```python
overtime_percentage = (df['OverTime'].value_counts(normalize=True) * 100)
print("Percentage of Employees Working Overtime:")
print(overtime_percentage)
```

```python
years_at_company_distribution = df['YearsAtCompany'].value_counts()
print("Years at Current Company Distribution:")
print(years_at_company_distribution)
```

```python
job_satisfaction_by_gender = df.groupby('Gender')['JobSatisfaction'].mean()
print("Job Satisfaction by Gender:")
print(job_satisfaction_by_gender)
```

```python
education_distribution = df['Education'].value_counts()
print("Education Level Distribution:")
print(education_distribution)
```

```python
department_counts = df['Department'].value_counts()
print("Department Employee Counts:")
print(department_counts)
```

# Data visualization

```python
numeric_data = df.select_dtypes(include='number')
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap for Numeric Variables")
plt.show()
```

```python
age_groups = ['<30', '30-39', '40-49', '50-59', '60+']
age_bins = [0, 29, 39, 49, 59, 120]
df['AgeGroup'] = pd.cut(df['ï»¿Age'], bins=age_bins, labels=age_groups)
plt.figure(figsize=(10, 6))
sns.countplot(x='AgeGroup', hue='OverTime', data=df)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Overtime by Age Group')
plt.legend(title='OverTime', labels=['No', 'Yes'])
plt.show()
```
